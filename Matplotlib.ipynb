# Matplotlib

Matplotlib es una de las librerías para visualización de datos en 2D más utilizadas en Python. Las gráficas que son posibles de generar pueden variar enormemente debido a la naturaleza de la información que se puede representar. Además, los elementos dentro de una gráfica como los ejes y los colores pueden ser manipulado para una mejor y más intuitiva lectura de resultados. Es por esto que esta librería es fundamental para áreas como ciencia de datos, matemáticas, biología, etc.

## Scales

Las escalas son importantes en Matplotlib porque nos permiten representar datos con diferentes rangos y magnitudes de manera adecuada. Al seleccionar la escala correcta para un gráfico, podemos resaltar patrones, tendencias y relaciones de manera más efectiva.

Por un lado podemos cambiar las unidades e intervalos como tal que se están usando para las figuras, pero también podemos cambiar la escala en relación a cuánto de la imagen mostramos por llamarlo de algún modo. En este último caso, imagina que tienes una imagen que va en el eje x desde -50 hasta 20, pero solo te interesa el intervalo -1:12, es ahí cuando podemos cambiar qué parte de la imagen vemos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo extraído de https://matplotlib.org/3.1.3/gallery/pyplots/pyplot_scales.html
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

### Change Y axes scale
# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear') # Assign the y scale
plt.title('linear y axis')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log') # Assign the y scale
plt.title('log y axis')
plt.grid(True)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
plt.show()

### Change X axes scale
plt.subplot(221)
plt.plot(x, y)
plt.xscale('linear') # Assign the x scale
plt.title('linear x axis')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.xscale('log') # Assign the x scale
plt.title('log x axis')
plt.grid(True)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
plt.show()
```

```python
# RECUERDA CORRER EL SCRIPT CADA QUE MODIFIQUES LOS PARÁMETROS
x = np.linspace(-50, 50, 100)
y = np.sin(x)

plt.plot(x, y)

xinf = -40 #@param {type:"slider", min:-50, max:100, step:1}
xsup = 45 #@param {type:"slider", min:-50, max:100, step:1}

yinf = -2 #@param {type:"slider", min:-2, max:0, step:1}
ysup = 2 #@param {type:"slider", min:1, max:15, step:1}

plt.xlim(xinf, xsup) # pone el límite inferior y superior en nuestro eje x
plt.ylim(yinf, ysup) # pone el límite inferior y superior en nuestro eje y

plt.show()
```

## Lines

Las líneas de las gráficas conectan diferentes vértices de nuestras imágenes y pueden tener diferentes estilos como colores, grosores y continuidades variadas. Por lo general solo se cambian el estilo de las líneas cuando estas no son visibles o se quiere marcar una distinción entre una línea y otra.

En este caso solo usaremos algunos de los posibles argumentos para distinguir algunas líneas en una misma gráfica. La figura y explicación original se puede observar en [este](https://matplotlib.org/stable/gallery/lines_bars_and_markers/line_demo_dash_control.html#sphx-glr-gallery-lines-bars-and-markers-line-demo-dash-control-py) sitio.

```python
# RECUERDA CORRER EL SCRIPT CADA QUE MODIFIQUES LOS PARÁMETROS

x = np.linspace(0, 10, 500)
y = np.sin(x)

plt.figure(figsize=(15, 4))

# Cambia el grosor de las líneas
width = 1.5 #@param {type:"slider", min:0.5, max:5, step:0.5}

plt.rc('lines', linewidth= width)
fig, ax = plt.subplots()

# Using set_dashes() and set_capstyle() to modify dashing of an existing line.
line1, = ax.plot(x, y, label='Usando set_dashes() y set_dash_capstyle()')
line1.set_dashes([2, 2, 10, 2])  # 2pt line, 2pt break, 10pt line, 2pt break.
line1.set_dash_capstyle('round')

# Using plot(..., dashes=...) to set the dashing when creating a line.
line2, = ax.plot(x, y - 0.2, dashes=[6, 2], label='Usando parámetro \"dashes\"')

# Using plot(..., dashes=..., gapcolor=...) to set the dashing and
# alternating color when creating a line.
# argument linestyle defines the shape of line
line3, = ax.plot(x, y - 0.4, linestyle = "dotted", gapcolor='tab:pink',
                 label='Uso de puntos y colores')

style = "solid" #@param ["dotted", "dashdot","dashed","solid"]
line4, = ax.plot(x, y - 0.6, linestyle = style, gapcolor='tab:green',
                 label='Uso de colores con estilo %s' %(style))

ax.legend(handlelength=4)
plt.show()
```

## Markers

Los *markers* se refieren a los marcadores utilizados para representar puntos en una gráfica. Los marcadores son símbolos visuales que se colocan en las coordenadas correspondientes a los datos en el gráfico. Estos marcadores pueden ser formas simples, como puntos, círculos, triángulos, cuadrados, etc., o incluso símbolos personalizados. Estos se utilizan comúnmente para resaltar o distinguir diferentes conjuntos de datos, o para enfatizar puntos específicos en un gráfico. Pueden ayudar a mejorar la legibilidad y comprensión de la información visualizada de modo que es necesario saber cuándo y cómo modificarlos.

En el ejemplo aquí presente se pueden observar algunos de los posibles símbolos para los marcadores habido en matplotlib; sin embargo, faltan varios. Si deseas ver la colección completa, puedes vicitarla en la [documentación](https://matplotlib.org/stable/api/markers_api.html) de la librería.

```python
# RECUERDA CORRER EL SCRIPT CADA QUE MODIFIQUES LOS PARÁMETROS

ypoints = np.array([3, 8, 1, 10])

mark = "o" #@param [".",",","o","^","<",">",1,2,3,4]

# cambia el tamaño del marcador
size = 21 #@param {type:"slider", min:1, max:50, step:1}

plt.plot(ypoints, marker = mark, linestyle = "dotted", ms = size)
```

## Colors y Colormaps

Los colores y las escalas de colores (*colormaps*) son fundamentales en Matplotlib porque desempeñan un papel crucial en la visualización de datos. Permiten agregar información visual atractiva y facilitar la interpretación de los gráficos generados. Permiten agregar información visual atractiva y facilitar la interpretación de los gráficos generados.

Los colores permiten diferenciar visualmente entre diferentes elementos o categorías en un gráfico. Al asignar colores distintos a conjuntos de datos o características específicas, podemos resaltar patrones, tendencias o relaciones en los datos. Además, los colores también pueden transmitir información adicional, como clasificaciones o niveles de intensidad como es el caso de los *heatmaps*.

Por otro lado, los *colormaps* proporcionan una forma de mapear valores numéricos a colores específicos. Al utilizar una escala de colores adecuada, podemos resaltar las variaciones y contrastes en los datos, lo que facilita la comprensión y la identificación de patrones.

```python
# Generate random data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Gráfica con diferentes colores
plt.figure(figsize=(8, 4))

plt.plot(x, y, color = 'blue', label = 'Azul')
plt.plot(x, y + 0.2, color = 'red', label = 'Rojo')
plt.plot(x, y + 0.4, color = 'green', label = 'Verde')
plt.plot(x, y + 0.6, color = 'purple', label = 'Morado')

plt.title('Grtáfica con varias colores')
plt.legend()
plt.show()

# Gráfica con colormap
plt.figure(figsize=(8, 4))

c = np.linspace(0, 1, 100)
plt.scatter(x, y, c=c, cmap='viridis')
plt.colorbar(label='Color')

plt.title('Gráfica con colormap')
plt.show()
```

## Legend

Las leyendas brindan información adicional sobre los elementos representados en un gráfico. Proporcionan una descripción clara de los elementos gráficos, como líneas, puntos o áreas, y permiten una mejor comprensión de los datos visualizados.

```python
# Random data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Graficar datos
plt.plot(x, y1, label='Sin(x)')
plt.plot(x, y2, label='Cos(x)')

# Manipular leyenda (loación, tamaño y sombra)
legend = plt.legend(loc='upper right', fontsize='small', shadow = True)

# Personalizar leyenda
legend.set_title('Trigonometric Functions')  # Poner el título
legend.get_title().set_fontsize('12')  # Tamaño del título de leyenda

# Efecto de sombra de la leyenda
legend.set_frame_on(True)
legend.get_frame().set_facecolor('lightgray')

# Bordes transparentes
legend.get_frame().set_alpha(0.8)

# Cambiar el color del texto de leyenda
for text in legend.get_texts():
    text.set_color('blue')

# Hacer el texto en negritas
for text in legend.get_texts():
    text.set_weight('bold')

# Mostrar la gráfica
plt.show()
```

# Ejemplos de plots

## Scatterplot

```python
# Datos random
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = np.random.randint(10, 100, size=100)

# scatter plot
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)

# personalización
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar(label='Color')

plt.show()
```

```python
from sklearn.datasets import load_iris

# Cargando la base de datos iris
iris = load_iris()
data = iris.data
target = iris.target

# Extraer información específica
x = data[:, 0]  # Sepal length
y = data[:, 1]  # Sepal width
colors = ['r', 'g', 'b']  # Color según la especie
species = iris.target_names[target]  # especie

# scatter plot
for i, species_name in enumerate(iris.target_names):
    species_indices = np.where(species == species_name)
    plt.scatter(x[species_indices], y[species_indices], c=colors[i], label=species_name, alpha=0.8)

# personalizar plot
plt.title('Iris Dataset Scatter Plot')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()

plt.show()
```

## Barplot

```python
# Clases
classes = ['Mammals', 'Birds', 'Fish', 'Reptiles', 'Amphibians']

# Abundancias
species_count = [5416, 10784, 33000, 10722, 8033]

# bar plot
plt.bar(classes, species_count)

# personalizar
plt.title('Number of Species per Biological Class')
plt.xlabel('Biological Class')
plt.ylabel('Number of Species')

plt.show()
```

```python
# Animales
species = ['Elephant', 'Giraffe', 'Lion', 'Dolphin', 'Turtle']

# población
population = [12, 8, 15, 6, 10]

# población máxima en reserva
max_population = [15, 10, 20, 8, 12]

# subplots
fig, ax = plt.subplots()

# graficar
ax.bar(species, population, label='Current Population', color = "red")
ax.bar(species, max_population, label='Maximum Sustainable Population', alpha=0.5)

# Poner límiite de eje y para máxima población
ax.set_ylim(0, max(max_population) * 1.2)

# Etiquetas y leyenda
for i in range(len(species)):
    ax.text(i, population[i] + 0.5, f"{population[i]}", ha='center', va='bottom')
    ax.text(i, max_population[i] + 0.5, f"{max_population[i]}", ha='center', va='bottom')

# personalizar
plt.title('Population of Animal Species in Wildlife Reserve')
plt.xlabel('Species')
plt.ylabel('Population')
plt.legend()

plt.show()
```

## Stackplot

```python
# Time points
time = np.arange(0, 10, 1)

# poblaciones de especies a lo largo de tiempo
sp1 = [50, 55, 60, 70, 80, 90, 85, 80, 75, 70]
sp2 = [30, 35, 40, 45, 50, 55, 60, 65, 70, 75]
sp3 = [20, 25, 30, 35, 40, 45, 50, 55, 60, 65]

# graficar
plt.stackplot(time, sp1, sp2, sp3, labels = ['Especie 1', 'Especie 2', 'Especie 3'])

plt.title('Distribución de poblaciones en el tiempo')
plt.xlabel('tiempo')
plt.ylabel('población')
plt.legend()

plt.show()
```
